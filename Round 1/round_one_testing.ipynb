```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'data/prices_round_1_day_-1.csv'

# Read the CSV file into a DataFrame, specifying the delimiter as semicolon
data = pd.read_csv(file_path, encoding='cp1252', delimiter=';')
print("Original DataFrame:")
print(data.head())
```

```python
data_starfruit = data[data["product"] == "STARFRUIT"]
price_starfruit = data_starfruit["mid_price"].values
price_starfruit = [float(s) for s in price_starfruit]
price_starfruit[:5]
```

```python
data_amethysts = data[data["product"] == "AMETHYSTS"]
price_amethysts = data_amethysts["mid_price"].values
price_amethysts = [float(s) for s in price_amethysts]
price_amethysts[:5]
```

```python
plt.plot(price_starfruit)
plt.xlabel('Time')
plt.ylabel('Price of starfruit')
plt.title('Price of starfruit on day -2')
plt.show()
```

```python
plt.plot(price_amethysts[:200])
plt.ylim(9990, 10010)
plt.xlabel('Time')
plt.ylabel('Price of amethysts')
plt.title('Price of amethysts on day -2')
plt.show()
```
