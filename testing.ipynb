```python
import json
import matplotlib.pyplot as plt
```

```python
with open('tutorial_data.json', 'r') as file:
    data_dict = json.load(file)

data_dict["prices"] = [price for price in data_dict["prices"] if price["symbol"] == "STARFRUIT"]
prices = [entry['price'] for entry in data_dict['prices']]
prices
```

```python
plt.plot(prices[:10], marker='o', linestyle='-')
plt.title('Price of STARFRUIT Over Time')
plt.xlabel('Time')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
