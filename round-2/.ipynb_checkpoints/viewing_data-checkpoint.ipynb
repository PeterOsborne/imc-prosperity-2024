```python
import json
with open('sandbox.json', 'r') as file:
    json_data = file.read()
parsed_data = json.loads(json_data)
lambda_logs = [item["lambdaLog"] for item in parsed_data["data"]]
```

```python
input_string = '========\n1059\n1060\n1067\n========\n1053\n1052\n1046\n========\n1055.0 1057.0\n0\nCONVERTING -1 ORCHIDS'
split_string = input_string.split('========\n')[1:]
best_bid = int(split_string[0][:4])
best_ask = int(split_string[1][:4])
split = split_string[2].split(' ')
conversion_bid = float(split[0])
conversion_ask = float(split[1][:6])
```

```python
def get_data(lamdba_logs):
    time_stamp = 0
    result = []
    for i in range(len(lambda_logs)):
        input_string = lambda_logs[i]
        split_string = input_string.split('========\n')[1:]
        best_bid = int(split_string[0][:4])
        best_ask = int(split_string[1][:4])
        split = split_string[2].split(' ')
        conversion_bid = float(split[0])
        conversion_ask = float(split[1][:6])
        result.append({
            "timestamp": time_stamp,
            "best_bid": best_bid,
            "best_ask": best_ask,
            "conversion_bid": conversion_bid,
            "conversion_ask": conversion_ask
        })
        time_stamp += 100
    return result
```

```python
data = get_data(lambda_logs)
```

```python
import matplotlib.pyplot as plt

timestamps = [entry['timestamp'] for entry in data]
best_bids = [entry['best_bid'] for entry in data]
best_asks = [entry['best_ask'] for entry in data]
conversion_bids = [entry['conversion_bid'] for entry in data]
conversion_asks = [entry['conversion_ask'] for entry in data]

plt.plot(timestamps[:100], best_bids[:100], label='Best Bid')
plt.plot(timestamps[:100], best_asks[:100], label='Best Ask')
plt.plot(timestamps[:100], conversion_bids[:100], label='Conversion Bid')
plt.plot(timestamps[:100], conversion_asks[:100], label='Conversion Ask')

plt.xlabel('Timestamp')
plt.ylabel('Values')
plt.title('Values vs Timestamp')
plt.legend()
plt.show()
```

```python
profit = 0
for i in range(len(data)):
    if best_bids[i] < conversion_bids[i]:
        profit += conversion_bids[i] - best_bids[i]
        print(f'buy market sell conversion at {timestamps[i]}')
    if conversion_asks[i] < best_asks[i]:
        print(f'buy conversion sell market at {timestamps[i]}')
```
